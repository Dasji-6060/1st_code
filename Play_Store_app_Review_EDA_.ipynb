## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to convert the Datatype.
play_store_app_review['Reviews']= pd.to_numeric(play_store_app_review['Reviews'])
play_store_app_review['Last Updated']= pd.to_datetime(play_store_app_review['Last Updated'])
play_store_app_review['Installs'] = play_store_app_review['Installs'].str.replace('[+,]', '', regex=True).astype(int)
````

````python
# Removing Duplicates
play_store_app_review.drop_duplicates(inplace=True)
play_store_app_review.shape
````

````python
# Filling missing Values with mean
play_store_app_review.Rating.fillna(play_store_app_review['Rating'].mean(), inplace=True)
play_store_app_review.Sentiment_Polarity.fillna(play_store_app_review['Sentiment_Polarity'].mean(), inplace=True)
play_store_app_review.Sentiment_Subjectivity.fillna(play_store_app_review['Sentiment_Subjectivity'].mean(), inplace=True)
````

````python
# Drop Missing Values
play_store_app_review.dropna()
````
